```python
import pyttsx3
from IPython.display import Audio

text = 'For the First Time In 18 Seasons, The ROYAL CHALLENGERS BANGALORE, Are the CHAMPIONS of The TATA IPL 2025. Thank you.'
audio = pyttsx3.init()
audio.setProperty('volume', 0.8) # Keep volume at a reasonable level
audio.setProperty('rate', 170) # Set a slightly slower rate for a calmer tone

voices = audio.getProperty('voices')

# --- Find a male voice ---
male_voice_found = False
selected_voice_name = "Default Voice" # To store the name of the selected voice

print("Available Voices and their properties:")
for i, voice in enumerate(voices):
    # Print all voice properties for debugging and selection
    gender_info = voice.gender if voice.gender else "N/A"
    name_info = voice.name if voice.name else "N/A"
    print(f"Voice {i}: ID='{voice.id}', Name='{name_info}', Gender='{gender_info}'")

    # Check for male voice, safely handling NoneType for gender and name
    if voice.gender and "male" in voice.gender.lower():
        audio.setProperty('voice', voice.id)
        selected_voice_name = voice.name if voice.name else voice.id # Use ID if name is None
        male_voice_found = True
        break
    elif voice.name and "david" in voice.name.lower(): # Common male voice on Windows
        audio.setProperty('voice', voice.id)
        selected_voice_name = voice.name if voice.name else voice.id
        male_voice_found = True
        break

if not male_voice_found:
    print("\nNo explicit male voice found based on gender or name keywords.")
    # Fallback: if no specific male voice is identified, try the first one
    # which is often male on many systems.
    if voices: # Ensure there's at least one voice
        try:
            audio.setProperty('voice', voices[0].id)
            selected_voice_name = voices[0].name if voices[0].name else voices[0].id
            print(f"Using default first voice: {selected_voice_name}")
        except IndexError:
            print("Error accessing default voice. No voices found.")
    else:
        print("No voices found on your system. Please ensure a TTS engine is installed.")
else:
    print(f"\nSelected Male Voice: {selected_voice_name}")

audio.say(text)
audio.runAndWait()
```
